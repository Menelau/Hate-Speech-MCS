```python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score
```

```python
def load_predictions(dataset_name):
    path = './Save Predict and Proba/' + dataset_name.upper() + '/pred_val_' + dataset_name.lower() + '.csv'
    table_pred = pd.read_csv(path)
    if dataset_name == 'zw':
        label = table_pred['norm']
        methods = table_pred.drop('norm', axis=1)
    else: 
        label = table_pred['class']
        methods = table_pred.drop('class', axis=1)
    methods = methods.drop('Unnamed: 0', axis=1)
    return label, methods
```

```python
label, methods = load_predictions(dataset_name)
```

```python
def get_oracle(labels, pred={}, clfs=[]):
    cont = 0;
    list_yes = np.zeros((len(labels)))
    oracle_preds = np.zeros((len(labels)))
    for i, k in enumerate(labels):
        acertou = False
        for clf in clfs:
            if pred[clf][i] == k:
                acertou = True
                oracle_preds[i] = pred[clf][i]
                break
        if acertou:
            cont += 1
            list_yes[i] = k
        else:
            oracle_preds[i] = pred[clf][0]
            if k == 0:
                list_yes[i] = 1
            else:
                list_yes[i] = 0
    return cont, list_yes, oracle_preds
```

```python
clfs = ['SVM-CV', 'SVM-TFIDF', 'SVM-W2V', 'SVM-GLOVE', 'SVM-FAST', 'KNN-CV', 'KNN-TFIDF',
        'KNN-W2V', 'KNN-GLOVE', 'KNN-FAST', ]
names = {}
names['td'] = ['MLP-FAST','KNN-GLOVE','KNN-FAST','KNN-CV','MLP-W2V','NB-CV','CNN-TFIDF',
               'NB-FAST','RF-CV','LR-CV','CNN-CV','KNN-W2V', 'CNN-GLOVE']

names['zw'] = ['CNN-W2V', 'MLP-CV', 'SVM-TFIDF','KNN-GLOVE','LR-FAST','MLP-GLOVE',
               'SVM-CV','CNN-CV','KNN-CV', 'MLP-FAST','RF-W2V','CNN-FAST', 'RF-FAST']

names['td_zw'] = ['EXTRA-CV','CNN-FAST','MLP-W2V','LR-CV','MLP-FAST','NB-CV','CNN-CV',
                  'LR-GLOVE','SVM-GLOVE','RF-FAST','RF-TFIDF','SVM-CV', 'KNN-CV']

oracle, df_yes, oracle_preds = get_oracle(label.to_numpy(), methods, names[dataset_name])
print(oracle)
```

```python
names = {}
names['fast'] = ['MLP-FAST','CNN-FAST','KNN-FAST','EXTRA-FAST','RF-FAST','SVM-FAST','LR-FAST','NB-FAST']
names['glove'] = ['MLP-GLOVE','CNN-GLOVE','KNN-GLOVE','EXTRA-GLOVE','RF-GLOVE','SVM-GLOVE','LR-GLOVE','NB-GLOVE']
names['cv'] = ['MLP-CV','CNN-CV','KNN-CV','EXTRA-CV','RF-CV','SVM-CV','LR-CV','NB-CV']
names['tfidf'] = ['MLP-TFIDF','CNN-TFIDF','KNN-TFIDF','EXTRA-TFIDF','RF-TFIDF','SVM-TFIDF','LR-TFIDF','NB-TFIDF']
names['w2v'] = ['MLP-W2V','CNN-W2V','KNN-W2V','EXTRA-W2V','RF-W2V','SVM-W2V','LR-W2V','NB-W2V']
                
for name in ['fast', 'glove', 'cv', 'tfidf', 'w2v']:
    oracle, df_yes, oracle_preds = get_oracle(label.to_numpy(), methods, names[name])
    print(f"{name}: {oracle}")
    score = f1_score(label.to_numpy(), oracle_preds, average='macro')
    print(f"{name} f1score: {score}")
```
